```python
import pandas as pd
import os
```

```python
folder = r'D:\FPTpoly\Khác\Projects\DWH_Sieu_Thi_Dien_May\DATA'

file_paths = [
    os.path.join(folder, sub_folder, file)
    for sub_folder in os.listdir(folder)
    if os.path.isdir(os.path.join(folder, sub_folder))
    for file in os.listdir(os.path.join(folder, sub_folder))
    if file.endswith('.xlsx')
]
```

```python
def replace_symbols(df, columns, old, new):
    for col in columns:
        df[col] = df[col].str.replace(old, new, regex=True)

def find_index(df, column_name):
    return list(df.columns).index(column_name) + 1

output_csv = r'D:\FPTpoly\Khác\Projects\DWH_Sieu_Thi_Dien_May\DATA\Tong_Hop_2020_-_2024.csv'

for file_path in file_paths:
    df = pd.read_excel(file_path, sheet_name='Sheet1')
    df_edit = df.copy()

    rename_map = {
        'Tên': 'Tên khách hàng',
        'Tên.1': 'Tên sản phẩm',
        'Mã': 'Mã khách hàng',
        'Mã.1': 'Mã đơn hàng',
        'Mã.2': 'Mã sản phẩm',
        'TT.1': 'Trạng thái'
    }
    df_edit.rename(columns=rename_map, inplace=True)

    drop_columns = ['Đối tác T/C', 'CT gốc', 'XK phụ kiện', 'TT']
    df_edit.drop(columns=drop_columns, inplace=True)

    address_columns = ['Tên khách hàng', 'Tên sản phẩm', 'Model', 'Tỉnh/TP', 'Q/Huyện', 'P/Xã', 'ĐC']
    replace_symbols(df_edit, address_columns[:6], ',', '.')
    replace_symbols(df_edit, [address_columns[6]], ',', '-')

    df_edit[address_columns] = df_edit[address_columns].fillna('None')

    desc_columns = ['0', '1', '2', '3']
    replace_symbols(df_edit, desc_columns, ',', '-')
    df_mo_ta = df_edit[desc_columns].apply(lambda x: ' '.join(x), axis=1)
    df_edit.drop(columns=desc_columns, inplace=True)

    product_index = find_index(df_edit, 'Tên sản phẩm')
    df_edit.insert(product_index, 'Mô tả sản phẩm', df_mo_ta)

    df_edit['Tiếp khách'].fillna('None', inplace=True)
    df_edit['NH'].fillna('None', inplace=True)

    df_edit.to_csv(output_csv, mode='a', index=False, header=not os.path.exists(output_csv))

print("Hoàn thành tổng hợp dữ liệu vào CSV!")
```
